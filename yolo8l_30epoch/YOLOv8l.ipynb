```python
from ultralytics import YOLO

model = YOLO("yolov8l.pt")

model.train(
    data='/content/yolo_dataset/dataset.yml',
    epochs=30,
    imgsz=640,
    batch=64,
    device=0,
    patience=10,
    name='deepfashion2_run_yolov8l_adamw',
    optimizer='AdamW',
    lr0=0.000588,
    momentum=0.9,
)
```
